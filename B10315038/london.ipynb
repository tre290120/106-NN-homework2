```python
cols = ['station_id', 'utc_time', 'PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']
x_need = ['station_id','PM2.5','PM10','NO2','temperature','pressure','humidity','wind_direction','wind_speed']
label_need = ['PM2.5', 'PM10']
```

```python
#初始化
all_df_x_train = pd.DataFrame(columns=x_need)
all_df_y_train = pd.DataFrame(columns=label_need)
all_df_x_test = pd.DataFrame(columns=x_need)
all_df_y_test = pd.DataFrame(columns=label_need)




aq_cols = ['utc_time', 'station_id', 'PM2.5', 'PM10', 'NO2']
all_df = all_df.values
all_df = pd.DataFrame(all_df, columns = aq_cols)
all_df['utc_time'] = all_df['utc_time'].astype(str)


meo_cols = ['stationName','longitude','latitude','utc_time','temperature','pressure','humidity','wind_direction','wind_speed']
all_meo = all_meo.values
all_meo = pd.DataFrame(all_meo, columns = meo_cols)
for sta in station_aq:
    #取得某個測站的所有資料
    get_sta = all_df[all_df['station_id']==sta]
    
    #get_sta = get_sta[663:]
    #print (get_sta)
    #處裡重複
    #aq_process = get_sta.drop_duplicates(subset=['station_id', 'utc_time', 'PM2.5', 'PM10', 'NO2'], keep='first')
    
    #取得對應的meo
    meo = all_meo[all_meo['stationName']==station_dic[sta]]
    #根據utc來做合併
    df = pd.merge(get_sta, meo, how='left', on='utc_time')
    
    rng = pd.date_range('1/1/2017', periods=9000, freq='H')
    time_dataframe=pd.DataFrame({'utc_time':rng})
    time_dataframe['utc_time'] = time_dataframe['utc_time'].astype(str)
    df = pd.merge(time_dataframe, df, how='left', on='utc_time')
    
    #將需要的col取出，並且轉成numpy array
    x = df[x_need]
    y = df[label_need]
    ndarray_x = x.values
    ndarray_y = y.values
    
    
    
    #判斷每個資料是不是完整的
    count = df.count(axis=1)
    filt_count = count[count==13]
    filt_index = filt_count.index
    
    #判斷是否是連續的data，並將其屬於第幾行存起來
    filter_x = []
    filter_y = []
    for index in range(len(filt_index)-1):
        if filt_index[index]+1 == filt_index[index+1]:
            filter_x.append(filt_index[index])
            filter_y.append(filt_index[index+1])
            
    
    
    
            
    ndarray_x_process = ndarray_x[filter_x]
    ndarray_y_process = ndarray_y[filter_y]
    x_dataframe=pd.DataFrame(ndarray_x_process, columns=x_need)
    y_dataframe=pd.DataFrame(ndarray_y_process, columns=label_need)
    

    msk = np.random.rand(len(x_dataframe)) < 0.8
    train_x_df = x_dataframe[msk]
    train_y_df = y_dataframe[msk]
    test_x_df = x_dataframe[~msk]
    test_y_df = y_dataframe[~msk]
    
    
    #train_x_df['wind_direction'] = train_x_df['wind_direction'].map({'999017':0}).astype(int)
    #train_x_df[train_x_df['wind_direction']==999017] = -1
    train_x_df['wind_direction'] = train_x_df['wind_direction'].replace(999017, -1)
    test_x_df['wind_direction'] = test_x_df['wind_direction'].replace(999017, -1)
    
    
    
    
    all_df_x_train = pd.concat([all_df_x_train, train_x_df],  ignore_index=True) 
    all_df_y_train = pd.concat([all_df_y_train, train_y_df],  ignore_index=True)
    all_df_x_test = pd.concat([all_df_x_test, test_x_df],  ignore_index=True)
    all_df_y_test = pd.concat([all_df_y_test, test_y_df],  ignore_index=True)
```

```python
x_train_OneHot_df = pd.get_dummies(data=all_df_x_train, columns=["station_id"])
x_test_OneHot_df = pd.get_dummies(data=all_df_x_test, columns=["station_id"])

train_Features = x_train_OneHot_df.values
y_temp = all_df_y_train.values
train_label_PM25, train_label_PM10 = y_temp[:,0], y_temp[:,1]

test_Features = x_test_OneHot_df.values
y_temp = all_df_y_test.values
test_label_PM25, test_label_PM10 = y_temp[:,0], y_temp[:,1]
```

```python
def show_train_history(train_history,train,validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()

    
modelPM25 = Sequential()
modelPM25.add(Dense(units=80, input_dim=19,
               kernel_initializer='uniform',
               activation='relu'))
modelPM25.add(Dense(units=60,
               kernel_initializer='uniform',
               activation='relu'))
modelPM25.add(Dense(units=30,
               kernel_initializer='uniform',
               activation='relu'))
modelPM25.add(Dense(units=1,
               kernel_initializer='uniform'))
modelPM25.compile(loss='mean_squared_error',
             optimizer='adam', metrics=['accuracy'])

#Training
train_historyPM25 = modelPM25.fit(x = train_Features,
                         y = train_label_PM25,
                         validation_split=0.1,
                         epochs=10,
                         batch_size=1,
                         verbose=2)
show_train_history(train_historyPM25, 'acc', 'val_acc')
show_train_history(train_historyPM25, 'loss', 'val_loss')

#Evaluate the training score
scoresPM25 = modelPM25.evaluate(x=test_Features,
                       y = test_label_PM25)
print (scoresPM25[1]) #print the accurity
```

```python
modelPM10 = Sequential()
modelPM10.add(Dense(units=80, input_dim=19,
               kernel_initializer='uniform',
               activation='relu'))
modelPM10.add(Dense(units=60,
               kernel_initializer='uniform',
               activation='relu'))
modelPM10.add(Dense(units=30,
               kernel_initializer='uniform',
               activation='relu'))
modelPM10.add(Dense(units=1,
               kernel_initializer='uniform'))
modelPM10.compile(loss='mean_squared_error',
             optimizer='adam', metrics=['accuracy'])

#Training
train_historyPM10 = modelPM10.fit(x = train_Features,
                         y = train_label_PM10,
                         validation_split=0.1,
                         epochs=10,
                         batch_size=1,
                         verbose=2)

scoresPM10 = modelPM10.evaluate(x=test_Features,
                       y = test_label_PM10)
print (scoresPM10[1]) #print the accurity
```
